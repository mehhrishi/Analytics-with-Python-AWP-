## Filtering 

```python
df.head()
```

```python
df.duration>=200
# it will give me the boolean values
# series form 
```

```python
# lets get the above output in the dataframe 
df[df.duration>=200]
```

```python
df[df.duration==200]
```

```python
df[df.title.sort_values]
```

```python
df[df.duration>=200]['title']
```

```python
df[df.duration>=200][['title']]
```

```python
df[df.duration>=200]['genre']
```

```python
df[df.duration>=200][['genre']]
```

```python
# fetching two columns
df[df.duration>=200][['genre','title']]
```

```python
# fetching more than two column or multiple colummns
df[df.duration>=200][['title','genre','duration']]
```

```python
# Using Logical Operators
(df.duration>=200) & (df.genre=='Crime')
# here will get boolean value
```

```python
df[(df.duration>=200) & (df.genre=="Crime")]
# it will give you all column depending upobn the condidtion you have put there
```

```python
# lets try to get only one Column 
df[(df.duration>=200) & (df.genre=="Action")][['title','genre']]
```

```python
df[(df.duration>=200) & (df.genre=="Action")]
```

```python
# OR operator "|"
df[(df.duration>=200) | (df.genre=='Crime')]
```

```python
df[(df.duration>=200) | (df.genre=='Crime') | (df.genre=='Action')]
```

```python
df[(df.duration>=200) | (df.genre=='Action')][['title','duration','genre']]
```

```python
df[(df.genre=="Crime") | (df.genre=="Adventure") | (df.genre=="Action")]
```
