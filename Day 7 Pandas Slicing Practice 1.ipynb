## Loc  & iLoc

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(8,4),
                 index=["a","b","c","d","e","f","g","h"],
                 columns=["A","B","C","D"])
df
```

## Loc - labelled Based Slicing

```python
df[['A','B']]
```

```python
df.loc[:,['A']]
```

```python
df.loc[:,['A','C']]
```

```python
df.loc[['a','c'],['A','C']]
```

```python
df.loc['a':'c','A':'C']
```

```python
df.loc[:,'A':'C']
```

```python
df.loc[::2,'A'::2]
```

```python
df.loc[["a","c"],["A","C"]]
```

```python
df.loc[:,"A":"C"]
```

```python
df.loc[['a'],:]
```

```python
df.loc["b":]
```

```python
df.loc['c':]
```

```python
df.loc['b':'f']
```

```python
df.loc['b':'f':2]
```

```python
df.loc["c":"f":2]
```

```python
df.loc[:,"A":]
```

```python
df.loc[:,"A"::2]
```

```python
df.loc[::2,::2]
```

## iLoc - Indexed Based Slicing

```python
df
```

```python
df.iloc[1:4,1:5]
```

```python
df.loc[1:4,1:5] 
# Loc doesnt support Index based Slicing
```

```python
df.iloc[1:4,1:]
```

```python
df.iloc[:,:-2]
```

```python
df.iloc[:-2,:]
```

```python
df.iloc[:-2,:]
```

```python
#Display first 2 rows and 2Cols

df.iloc[:2,:2]
```

```python
#display rows and col step of 2
df.iloc[::2,::2]
```
