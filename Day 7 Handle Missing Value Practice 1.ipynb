```python
import numpy as np 
import pandas as pd
```

```python
data = {
    'Name':['Tom','James','Ricky','Messi','Ronaldo',np.nan,'Jack'],
    'Age':[25,26,np.nan,23,np.nan,29,23],
    'Rating':[4.23,3.24,3.98,np.nan,4.6,np.nan,4.32]
}

df = pd.DataFrame(data)
df
```
