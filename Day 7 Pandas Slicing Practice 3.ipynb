## Loc - Label Based Slicing

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(8,4),
                 index=['a','b','c','d','e','f','g','h'],
                 columns=['A','B','C','D'])
df
```

```python
df[['A','B']]
```

### iLoc - Index Based Slicing

```python
df
```

```python
df.iloc[:4,:4]
```

```python
df.iloc[:4,[0,2,3]]
```

```python
df.iloc[1:4,1:4]
```

```python
df.iloc[1:4,1:5]
```

```python
# Loc doesnt support index Based Slicing 
df.loc[1:4,1:4]
```

```python
df.iloc[::2,:]
```

```python
df.iloc[::2]
```

```python
df.iloc[:,::1]
```

```python
df.iloc[1:4,1:]
```

```python
df.iloc[:-2,:]
```

```python
df.iloc[::-1,::-1]
```

```python
# Display first 2 rows and 2 Cols
df.iloc[:2,:2]
```

```python
# Display last two rows and cols
df.iloc[-2:,-2:]
```

```python
#display rows and col step of 2
df.iloc[::2,::2]
```
