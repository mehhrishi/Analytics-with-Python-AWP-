# Loc - Label Based Slicing 
**Endpoint is Inclusive**
    [Rows and Columns] 

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(8,4),
                 index=['a','b','c','d','e','f','g','h'],
                 columns=['A','B','C','D'])

df
```

```python
df[['A','B']]
```

```python
# it will extract all the rows
# it will extract only A and B Column 
df.loc[:,["A","B"]]
```

```python
df.loc[::2,["A","B"]]
```

```python
df.loc[::2,::2]
```

```python
# Selecting particular column from the data frame
df.loc[['a','c','e'],['A','B']]
```

```python
df.loc[['a'::2],['A','B']]
```

```python
df.loc[::2,::2]
```

```python
df.loc['a'::2,'A'::2]
```

```python
df.loc['a':"c","A":"C"] 
```

```python
df.loc[:,'A':"C"]
```

```python
df.loc[::2,'A'::2]
```

```python
df.loc[["a","c"],["A","C"]]
```

```python
df.loc[['a','c'],['A','C']]
```

```python
df.loc[:,'A':'C']
```

```python
df.loc[['a'],:]
```

```python
df.loc["b":]
```

```python
df.loc['b':,:]
```

```python
df.loc[["b":],:]
```

```python
df.loc["c":]
```

```python
df.loc["b":"f"]
```

```python
df.loc['b':'f':2]
```

```python
df.loc["A":"D"]
```

```python
df.loc['a'::2,'A'::2]
```

```python
df.loc[:,"A":]
```

```python
df.loc[:,"A"::2]
```

## iLoc - Index Based Slicing

```python
df 
```

```python
df.iloc[:4,:4]
```

```python
df.iloc[1:4,1:4]
```

```python
df.iloc[1:4,1:5]
```

```python
# Loc doesnt support Index Based Slicing
df.loc[1:4,1:5]
```

```python
df.iloc[::2,:]
```

```python
df.iloc[:,::1]
```

```python
df.iloc[1:4,1:]
```

```python
df.iloc[:-2,:]
```

```python
df.iloc[::-1,::-1]
```

```python
#Display first 2 rows and 2Cols
df.iloc[:2,:2]
```

```python
# Displaty last two rowas and columns
df.iloc[-2::1,-2::1]
```

```python
#display rows and col step of 2
df.iloc[::2,::2]
```
